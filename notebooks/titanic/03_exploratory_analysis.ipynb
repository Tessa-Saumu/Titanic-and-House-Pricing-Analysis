### 1. Univariate Analysis
**Goal:** Understand the distribution of individual features and the baseline survival rate.

```python
# Baseline Survival Rate
plot_bar_chart(df, 'Survived', "Overall Survival Rate (0 = No, 1 = Yes)")

# Distribution of Age and Fare
plot_histogram(df, 'Age', "Distribution of Passenger Age")
plot_histogram(df, 'Fare', "Distribution of Passenger Fare")
```

### 2. Bivariate Analysis
**Goal:** Determine how key features relate to our target variable (`Survived`).
We are looking for clear separation between survivors and non-survivors.

```python
# Survival by Sex
plot_bar_chart(df, 'Sex', "Survival Count by Gender") # Using Sex as a proxy for survival patterns

# Survival by Class
plot_box(df, 'Pclass', 'Survived', "Survival Distribution by Passenger Class")

# Survival by Age (using Boxplot to see medians)
plot_box(df, 'Pclass', 'Age', "Age Distribution by Pclass")
```

### 3. Correlation Analysis
**Goal:** Identify linear relationships between numerical features.

```python
plot_correlation_heatmap(df, "Feature Correlation Heatmap")
```
