# Phase 3: Data Cleaning (Titanic)

**Objective:** Execute the cleaning plan formulated in Phase 2 to prepare the dataset for Exploratory Data Analysis (EDA).

Based on our profiling, our strategy is:
1. **Drop `Cabin`**: 77% missing data.
2. **Drop `Embarked` nulls**: Negligible data loss (2 rows).
3. **Preserve `Age` nulls**: We will defer imputation to the machine learning pipeline (Week 4) to prevent **data leakage**. Pandas handles `NaN` values gracefully during EDA.

### 1. Environment Setup
We load our custom configuration and data ingestion tools. We use `autoreload` to ensure that any changes we make to our `src/` modules are instantly reflected in this notebook without needing to restart the kernel.

```python
%load_ext autoreload
%autoreload 2
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent.parent))

from src.data.loader import load_raw_data, save_processed_data
```

### 1.1 Load the Data
We load our data using our reusable loader function

```python
df_titanic = load_raw_data("Titanic.csv")
```

### 2. Execution & Defensive Checks
We will apply our cleaning steps and immediately follow them with `assert` statements. This is a defensive programming best practice—it ensures our script fails loudly if the data doesn't behave exactly as we expect, preventing silent errors downstream.

```python
# 1. Drop Cabin column
df_titanic = df_titanic.drop(columns=['Cabin'])

# 2. Drop rows where Embarked is missing
df_titanic = df_titanic.dropna(subset=['Embarked'])

# 3. Defensive Checks (Assertions)
assert 'Cabin' not in df_titanic.columns, "Cabin column was not dropped!"
assert df_titanic['Embarked'].isnull().sum() == 0, "Nulls remain in Embarked!"
assert df_titanic['Age'].isnull().sum() > 0, "Age nulls were accidentally dropped!"
assert len(df_titanic) == 889, f"Expected 889 rows, got {len(df_titanic)}"

print("Data cleaning steps executed and verified successfully.")
```
