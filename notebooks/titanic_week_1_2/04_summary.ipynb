# Phase 5: Summary & Model Readiness (Titanic)

## 1. Insights
- **Key Drivers:** `Sex` and `Pclass` are the strongest categorical predictors of survival.
- **Demographic Patterns:** `Age` exhibits bimodality (a peak in toddlers), indicating age-based survival advantages. 

## 2. Limitations
- **Missing Data Loss:** The `Cabin` feature was dropped entirely due to 77% missing data, potentially losing spatial/location-based survival signals.
- **Data Leakage Risk:** `Age` is missing ~20% of its values. Imputation was deliberately deferred to avoid contaminating the EDA and test sets.

## 3. Assumptions
- Features like `Name` and `Ticket` are assumed to lack baseline predictive power without complex Natural Language Processing (NLP) engineering, which is out of scope for the initial model.

## 4. Model Readiness (Week 4 Requirements)
To prepare for Machine Learning, the following preprocessing pipeline steps are required:
1. **Imputation:** `Age` missing values must be imputed using the *training set median*.
2. **Encoding:** `Sex`, `Embarked`, and `Pclass` require One-Hot Encoding.
3. **Feature Dropping:** Drop `PassengerId`, `Name`, and `Ticket` before training to prevent overfitting on identifiers.
