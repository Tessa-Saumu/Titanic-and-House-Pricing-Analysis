# Phase 5: Summary & Model Readiness (Housing)

## 1. Insights
- **Key Drivers:** `Area` and `Bathrooms` are the strongest numerical predictors, while `Airconditioning` is the strongest categorical driver.
- **Distributions:** Both the target (`Price`) and primary feature (`Area`) are heavily right-skewed.

## 2. Limitations
- **Sample Size Variance:** High-end categories (e.g., 4 bathrooms) have insufficient sample sizes, which may cause models to struggle with high-end price predictions.
- **Outliers:** The presence of extreme outliers in `Price` may negatively impact distance-based or linear models.

## 3. Assumptions
- We assume the linear relationships observed in EDA (e.g., `Area` vs. `Price`) will generalize to unseen data.

## 4. Model Readiness (Week 4 Requirements)
While structurally clean (0 missing values), the dataset requires strict mathematical transformations:
1. **Binary Mapping:** Yes/No categorical strings (`mainroad`, `airconditioning`, etc.) must be mapped to 1/0 integers.
2. **Encoding:** Multi-class strings (`furnishingstatus`) require One-Hot Encoding.
3. **Feature Scaling:** Due to massive scale discrepancies (e.g., `Area` in the thousands vs. `Bathrooms` in single digits), numerical features must pass through a `StandardScaler` or `MinMaxScaler`.
